pie charts and bar charts

```python
import numpy as np
import matplotlib.pyplot as plt
```

pie function api

```python
labels = 'frogs', 'hogs', 'dogs', 'logs'
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)

plt.pie(x=sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')

plt.show()
```

bar chart

```python
N = 5
menMeans = (20, 35, 30, 35, 27)
menStd = (2, 3, 4, 1, 2)

ind = np.arange(N)
width = 0.35

fig, ax = plt.subplots()
rectangles1 = ax.bar(ind, menMeans, width, color='r', yerr=menStd)

womenMeans = (25, 32, 34, 20, 25)
womenStd = (3, 5, 2, 3, 3)
rectangles2 = ax.bar(width+ind, womenMeans, width, color='y', yerr=womenStd)


ax.set_ylabel('scores')
ax.set_title('scores by group and gender')

ax.set_xticks(width/2+ind)
ax.set_xticklabels(('g1', 'g2', 'g3', 'g4', 'g5'))

ax.legend((rectangles1[0], rectangles2[0]), ('men', 'women'))
```
