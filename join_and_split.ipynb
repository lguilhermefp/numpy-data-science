```python
arrays = np.zeros((5, 3, 4))
for n in range(5):
    arrays[n] = np.random.randn(3, 4)

arrays
```

```python
stack0 = np.stack(arrays, axis=0)
stack1 = np.stack(arrays, axis=1)
stack2 = np.stack(arrays, axis=2)
```
